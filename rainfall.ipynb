# USING MULTIPLE LINEAR REGRESSION FOR RAINFALL PREDICTION TO AID IN IMPROVING ACCURACY OF WEATHER FORECAST

# Importing our libraries
1. Pandas
2. Sklearn
3. Math

```python
import pandas
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

# Dataset before cleaning

```python
df = pandas.read_csv("weather.csv")
df
```

# Dataset after cleaning

```python
df = pandas.read_csv("file.csv")
df
```

# Defining our data attributes. Independent variables
They include:
1. Evaporation
2. Sunshine
3. WindSpeed
4. Humidity
5. Pressure
6. Temperature

```python
x = df[['Evaporation', 'Sunshine', 'WindSpeed', 'Humidity', 'Pressure', 'Temperature']]
x
```

# Defining our dependent variable
1. Rainfall.
This is the attribute that our model will try and predict from the data

```python
y = df['Rainfall']
y
```

# Splitting our data into training and testing sets.
The data is split into:
1. 80% for training
2. 20% for testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

# Initializing our Linear regresion model
We initialize our model by creating an instance of the class LinearRegression

```python
regr = linear_model.LinearRegression()
```

# Training the model
This is done by calling the fit() method which is given the data inputs, x_train and y_train as arguments

```python
regr.fit(x_train, y_train)
```

# Testing the model
the model is tested by calling the predict() method which is given data input, x_test as arguments

```python
y_prediction = regr.predict(x_test)
```

# Printing the model perfomance metrics
 We print the mean squared error(MSE) which is the Mean or Average of the square of the difference between actual and estimated values. it is squared since some eroor values are negative. we then find the square root of MSE to find the mean error.

```python
mse = mean_squared_error(y_test, y_prediction)
error = sqrt(mse)
print(f"mean squared error is {mse}")
print(f"error value is {error}")
```
